```python
import yfinance as yf
import mplfinance as mpf
import pandas as pd

# Fetch recent data
ticker = yf.Ticker("BTC-USD")
minute_data = ticker.history(period="max", interval="1m")

# Plot
mpf.plot(minute_data, type='candle', volume=True, style='yahoo')
# print('hello')
```
